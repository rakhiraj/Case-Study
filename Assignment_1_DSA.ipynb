#### 5. Perform meaningful visualizations using the dataset. Bring at least 3 visualizations.

```python
sns.countplot(data["Classification"])
```

```python
plt.hist(data["SL"],color="purple",bins=20)
plt.title("Histogram of Sepal Length")
plt.xlabel("SL")
plt.ylabel("No. of observations")
```

```python
plt.scatter(data["SL"],data["PL"],s=15)
plt.title("Relationship between sepal length and petal width")
plt.xlabel("SL")
plt.ylabel("PL")
```

```python
sns.scatterplot(data["PL"],data["PW"],hue=data["Classification"]).set(title="Scatterplot of petal length and petal width")
```

```python
sns.scatterplot(data["SL"],data["SW"],hue=data["Classification"]).set(title="Scatterplot of sepal length and sepal width")
```

```python
sns.boxplot(x=data["PW"],y=data["SW"],hue=data["Classification"]).set(title="Boxplot of petal width and sepal width")
```
