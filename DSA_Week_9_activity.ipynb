### 7. Check and handle missing values in the dataset.

```python
#checking null values
df.isna().sum()
```

```python
#filling missing values of column age with mean values
avg=df["age"].mean()
df["age"]=df["age"].fillna(avg)
df
```

```python
#filling missing values of column income with median values
med=df["income"].median()
df["income"]=df["income"].fillna(med)
df
```

```python
#filling missing values of column gender with mode values
mod=df["gender"].mode()[0]
df["gender"]=df["gender"].fillna(mod)
df
```

### 8. Check the outliers and handle outliers in performance score using Percentiles.

```python
df["performance_score"].describe()
```

```python
plt.boxplot(df["performance_score"])
plt.title("Boxplot of Performance Score",fontsize=15)
```

```python
#finding 1st quantile
q1=np.percentile(df["performance_score"],25)
#findind 3rd quantile
q3=np.percentile(df["performance_score"],75)
#finding inter quartile range
IQR=q3-q1
#finding the lower limit
low_lim=q1-1.5*IQR
#finding the outlier
outlier=df[df["performance_score"]<low_lim]
#getting the index of outlier
outlier.index
```
