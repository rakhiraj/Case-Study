### 2. Try out different clustering models in the wine dataset.
### 3. Find the optimum number of clusters in each model and create the model with the optimum number of clusters.

### Scaling data

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(df)
```

### K-Means Clustering

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 10):
    kmeans = KMeans(n_clusters = i , init = 'k-means++', random_state = 21)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

```python
plt.figure(figsize = (10,8))
plt.plot(range(1,10), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

#### Silhouette Validation

```python
from sklearn.metrics import silhouette_score
score=[]
plt.figure(figsize=(10,10))
for i in range(2,10):
    k=KMeans(i)
    k.fit(x)
    score.append(np.round(silhouette_score(x,k.labels_),2))
plt.plot(range(2,10),score)
plt.show()
```

Here we can see 3 gives the highest score, so we go for k value 3(k=3).

```python
kmeans = KMeans(n_clusters = 3, init = 'k-means++', random_state = 42)
```

```python
kmeans.fit(x)
```

```python
cluster_labels = kmeans.fit_predict(x)
```

```python
print(cluster_labels)
```

```python
df1=df.copy()
df1['k_cluster'] = cluster_labels
df1.head()
```

```python
df1['k_cluster'].value_counts()
```

```python
df1.groupby('k_cluster').mean()
```

### Agglomerative Clustering

```python
import scipy.cluster.hierarchy as sch
plt.figure(figsize = (20,8))
dendrogram = sch.dendrogram(sch.linkage(x, method = 'ward'))
plt.title('Dendrogram')
plt.ylabel('Euclidean distances')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
score=[]
for i in range(2,10):
    hc=AgglomerativeClustering(i)
    hc.fit(x)
    score.append(np.round(silhouette_score(x,hc.labels_),2))
plt.plot(range(2,10),score)
plt.show()
```

Here we got 3 is good for k (cluster)

```python
res = AgglomerativeClustering(n_clusters = 3, affinity = 'euclidean', linkage = 'ward')
labels = res.fit_predict(x)
print(labels)
```

```python
df1['agg_cluster'] = labels
df1.head()
```

```python
df1.groupby('agg_cluster').mean()
```
